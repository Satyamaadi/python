```
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```
def show_with_matplotlib(image,title,pos):
  img_RGB = image[:,:,::-1]
  ax = plt.subplot(2,3,pos)
  plt.imshow(img_RGB)
  plt.title(title)
  plt.axis('off')
```

```
fig = plt.figure(figsize=(15, 7))
plt.suptitle("Adaptive thresholding", fontsize=14, fontweight='bold')
fig.patch.set_facecolor('silver')
```

```
!wget https://upload.wikimedia.org/wikipedia/commons/thumb/f/ff/Sudoku-by-L2G-20050714.svg/1200px-Sudoku-by-L2G-20050714.svg.png
```

```
image = cv2.imread('1200px-Sudoku-by-L2G-20050714.svg.png')
```

```
gray_image = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
```

```
thresh1 = cv2.adaptiveThreshold(gray_image,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,11,2)
```

```
show_with_matplotlib(cv2.cvtColor(thresh1, cv2.COLOR_GRAY2BGR), "method=THRESH_MEAN_C, blockSize=11, C=2", 2)
```

```
thresh2 = cv2.adaptiveThreshold(gray_image,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,31,3)
```

```
show_with_matplotlib(cv2.cvtColor(thresh2,cv2.COLOR_GRAY2BGR),'method=THRESH_MEAN_C, blozksize=31, C=3',3)
```

```
thresh3 = cv2.adaptiveThreshold(gray_image, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 2)
thresh4 = cv2.adaptiveThreshold(gray_image, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 31, 3)
```

```
show_with_matplotlib(cv2.cvtColor(thresh3, cv2.COLOR_GRAY2BGR), "method=GAUSSIAN_C, blockSize=11, C=2", 5)
show_with_matplotlib(cv2.cvtColor(thresh4, cv2.COLOR_GRAY2BGR), "method=GAUSSIAN_C, blockSize=31, C=3", 6)
```
