```
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```
def show_with_matplotlib(color_img, title, pos):
    """Shows an image using matplotlib capabilities"""

    # Convert BGR image to RGB
    img_RGB = color_img[:, :, ::-1]

    ax = plt.subplot(3, 4, pos)
    plt.imshow(img_RGB)
    plt.title(title)
    plt.axis('off')


# Create the dimensions of the figure and set title:
plt.figure(figsize=(12, 6))
plt.suptitle("Comparing different kernels using cv2.filter2D()", fontsize=14, fontweight='bold')
```

```
image = cv2.imread('_111434467_gettyimages-1143489763.jpg')
```

```
type(image)
```

```
plt.imshow(image)
```

```
kernel_identity = np.array([[0,0,0],[0,1,0],[0,0,0]])
```

```
original_image = cv2.filter2D(image,-1,kernel_identity)
```

```
plt.imshow(original_image)
```

```
kernel_edge_detection = np.array([[1,0,-1],[0,0,0],[-1,0,1]])
```

```
edge_detection_1 = cv2.filter2D(image,-1,kernel_edge_detection)
```

```
plt.imshow(edge_detection_1)
```

```
kernel_edge_detection_2 = np.array([[0,1,0],[1,-4,1],[0,1,0]])
```

```
edge_detection_2 = cv2.filter2D(image,-1,kernel_edge_detection_2)
```

```
plt.imshow(edge_detection_2)
```

```
kernel_sharpen = np.array([[0,-1,0],[-1,5,-1],[0,-1,0]])
```

```
sharpen_image = cv2.filter2D(image,-1,kernel_sharpen)
```

```
plt.imshow(sharpen_image)
```

```
kernel_unsharp_mask = -1/256.0 * np.array([[1,4,6,4,1],[4,16,24,16,4],[6,24,-476,24,6],[4,16,24,16,4],[1,4,6,4,1]])
```

```
unsharp_mask_image = cv2.filter2D(image,-1,kernel_unsharp_mask)
```

```
plt.imshow(unsharp_mask_image)
```

```
kernel_blur = 1/9 * np.array([[1,1,1],[1,1,1],[1,1,1]])
```

```
blur_image_1 = cv2.filter2D(image,-1,kernel_blur)
```

```
plt.imshow(blur_image_1)
```

```
gaussian_blur = 1/16*np.array([[1,2,1],[2,4,2],[1,2,1]])
```

```
blur_image_2 = cv2.filter2D(image,-1,gaussian_blur)
```

```
plt.imshow(blur_image_2)
```

```
kernel_emboss = np.array([[-2,-1,0],[-1,1,1],[0,1,2]])
```

```
emboss_image = cv2.filter2D(image,-1,kernel_emboss)
```

```
plt.imshow(emboss_image)
```

```
sobel_x_kernel = np.array([[1, 0, -1],[2, 0, -2],[1, 0, -1]])
```

```
sobel_image = cv2.filter2D(image,-1,sobel_x_kernel)
```

```
plt.imshow(sobel_image)
```

```
sobel_y_kernel = np.array([[1, 2, 1],[0, 0, 0],[-1, -2, -1]])
```

```
sobel_image_1 = cv2.filter2D(image,-1,sobel_y_kernel)
```

```
plt.imshow(sobel_image_1)
```

```
outline_kernel = np.array([[-1, -1, -1],[-1, 8, -1],[-1, -1, -1]])
```

```
outlined_image = cv2.filter2D(image,-1,outline_kernel)
```

```
plt.imshow(outlined_image)
```
