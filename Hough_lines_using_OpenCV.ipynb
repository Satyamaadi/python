```
import cv2
import numpy as np
```

```
!wget https://raw.githubusercontent.com/PacktPublishing/Learning-OpenCV-4-Computer-Vision-with-Python-Third-Edition/master/chapter03/lines.jpg
```

```
image = cv2.imread('lines.jpg')
```

```
type(image)
```

```
gray_image = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
```

```
edges = cv2.Canny(gray_image,50,120)
```

```
minLineLength = 20
maxLineGap = 5
```

```
lines = cv2.HoughLinesP(edges,1,np.pi/180.0,20,minLineLength,maxLineGap)
```

```
for x1, y1, x2, y2 in lines[0]:
    cv2.line(image, (x1, y1), (x2, y2), (0,255,0),2)
```

```
lines = lines.astype('uint8')
```
