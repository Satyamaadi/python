```
import cv2
import numpy as np
```

```
!wget https://raw.githubusercontent.com/Apress/building-computer-vision-apps-artificial-neural-networks/master/chapter3/images/soccer-in-green.jpg
```

```
image = cv2.imread('soccer-in-green.jpg')
```

```
translation_matrix = np.float32([[1,0,50],[1,0,20]])
```

```
movedImage = cv2.warpAffine(image,translation_matrix,(image.shape[0],image.shape[1]))
```

```
import matplotlib.pyplot as plt
```

```
plt.imshow(movedImage)
```
