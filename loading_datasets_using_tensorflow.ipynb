```
import tensorflow_datasets as tfds
```

```
ds,info = tfds.load('cats_vs_dogs',split='train',with_info=True)
```

```
ds
```

```
ds = ds.take(1)
for example in ds:
  print(list(example.keys()))
  image = example['image']
  label = example['label']
  print(image.shape,label)
```

```
tfds.as_dataframe(ds.take(4), info)
```

```
fig = tfds.show_examples(ds, info)
```
