#Skin segementation

```
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```
!wget https://www.accenture.com/t20200622T032553Z__w__/us-en/_acnmedia/Accenture/Redesign-Assets/DotCom/Images/Global/General/54/Accenture-Jim-Wilson-724x543.png
```

```
lower_hsv = np.array([0,40,80],dtype='uint8')
upper_hsc = np.array([20,255,255],dtype='uint8')
```

```
image = cv2.imread('Accenture-Jim-Wilson-724x543.png')
```

```
type(image)
```

```
hsv_image = cv2.cvtColor(image,cv2.COLOR_BGR2HSV)
```

```
plt.imshow(hsv_image)
```

```
skin_detected = cv2.inRange(hsv_image,lower_hsv,upper_hsc)
```

```
plt.imshow(skin_detected)
```
