```
import cv2
import numpy as np
from scipy import ndimage
```

```
kernel_3x3 = np.array([[-1, -1, -1],
                       [-1,  8, -1],
                       [-1, -1, -1]])

kernel_5x5 = np.array([[-1, -1, -1, -1, -1],
                       [-1,  1,  2,  1, -1],
                       [-1,  2,  4,  2, -1],
                       [-1,  1,  2,  1, -1],
                       [-1, -1, -1, -1, -1]])
```

```
!wget https://raw.githubusercontent.com/PacktPublishing/Learning-OpenCV-4-Computer-Vision-with-Python-Third-Edition/master/images/statue_small.jpg
```

```
image = cv2.imread('statue_small.jpg',0)
```

```
k3 = ndimage.convolve(image, kernel_3x3)
k5 = ndimage.convolve(image, kernel_5x5)
```

```
blurred = cv2.GaussianBlur(image, (17,17), 0)
g_hpf = image - blurred
```

```
import matplotlib.pyplot as plt
```
