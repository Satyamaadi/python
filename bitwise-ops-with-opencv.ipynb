```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def show_with_matplotlib(color_img,title,pos):
    img_rgb = color_img[:,:,::-1]
    ax = plt.subplot(3,3,pos)
    plt.imshow(img_rgb)
    plt.axis('off')
    plt.title(title)
    
plt.figure(figsize=(12,6))
plt.suptitle('Bitwise op with open cv 4',fontsize=14,fontweight='bold')

```

```python
img1 = np.zeros((300,300),dtype='uint8')
cv2.rectangle(img1,(10,10),(110,110),(255,255,255),-1)
cv2.circle(img1,(200,200),50,(255,255,255),-1)
```

```python
img2 = np.zeros((300,300),dtype='uint8')
cv2.rectangle(img2,(50,50),(150,150),(255,255,255),-1)
cv2.circle(img2,(225,200),50,(255,255,255),-1)
```

```python
bitwise_or = cv2.bitwise_or(img1,img2)
```

```python
bitwise_and = cv2.bitwise_and(img1,img2)
```

```python
bitwise_xor = cv2.bitwise_xor(img1,img2)
```

```python
bitwise_not = cv2.bitwise_not(img1)
```

```python
show_with_matplotlib(cv2.cvtColor(img1, cv2.COLOR_GRAY2BGR), "image 1", 1)
show_with_matplotlib(cv2.cvtColor(img2, cv2.COLOR_GRAY2BGR), "image 2", 2)
show_with_matplotlib(cv2.cvtColor(bitwise_or, cv2.COLOR_GRAY2BGR), "image 1 OR image 2", 3)
show_with_matplotlib(cv2.cvtColor(bitwise_and, cv2.COLOR_GRAY2BGR), "image 1 AND image 2", 4)
show_with_matplotlib(cv2.cvtColor(bitwise_xor, cv2.COLOR_GRAY2BGR), "image 1 XOR image 2", 5)
show_with_matplotlib(cv2.cvtColor(bitwise_not, cv2.COLOR_GRAY2BGR), "NOT (image 1)", 6)
#show_with_matplotlib(cv2.cvtColor(bitwise_not_2, cv2.COLOR_GRAY2BGR), "NOT (image 2)", 7)
```
