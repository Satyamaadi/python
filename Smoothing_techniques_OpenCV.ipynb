```
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```

def show_with_matplotlib(color_img, title, pos):
    """Shows an image using matplotlib capabilities"""

    # Convert BGR image to RGB
    img_RGB = color_img[:, :, ::-1]

    ax = plt.subplot(1, 1, pos)
    plt.imshow(img_RGB)
    plt.title(title)
    plt.axis('off')


# Create a figure() object with appropriate size and title
plt.figure(figsize=(12, 6))
plt.suptitle("Smoothing techniques", fontsize=14, fontweight='bold')
```

```
image = cv2.imread('02-cat-training-NationalGeographic_1484324.ngsversion.1526587209178.adapt.1900.1.jpg')
```

```
type(image)
```

```
plt.imshow(image)
```

```
kernel_averaging = np.ones((10,10),np.float32)/100
```

```
kernel_averaging_5_5 = np.array([[0.04, 0.04, 0.04, 0.04, 0.04],
                                 [0.04, 0.04, 0.04, 0.04, 0.04],
                                 [0.04, 0.04, 0.04, 0.04, 0.04],
                                 [0.04, 0.04, 0.04, 0.04, 0.04],
                                 [0.04, 0.04, 0.04, 0.04, 0.04]])
```

```
smooth_image_1 = cv2.filter2D(image,-1,kernel_averaging_5_5)
```

```
smooth_image_2 = cv2.filter2D(image,-1,kernel_averaging)
```

```
show_with_matplotlib(smooth_image_1,'Averaging kernel',1)
show_with_matplotlib(smooth_image_2,'10 by 10 kerner',1)
```

```
smooth_image_3 = cv2.blur(image,(10,10))
```

```
show_with_matplotlib(smooth_image_3,'Using blur fn',1)
```

```
smooth_image_4 = cv2.GaussianBlur(image,(9,9),0)
```

```
show_with_matplotlib(smooth_image_4,'Gaussian blur',1)
```

```
smooth_image_5 = cv2.medianBlur(image,9)
```

```
show_with_matplotlib(smooth_image_5,'Median Blur',1)
```

```
smooth_image_bf = cv2.bilateralFilter(image, 5, 10, 10)
smooth_image_bf_2 = cv2.bilateralFilter(image, 9, 200, 200)
```

```
show_with_matplotlib(smooth_image_bf,'Bilateral filter',1)
```

```
show_with_matplotlib(smooth_image_bf_2,'2nd Bilateral Filter',1)
```
