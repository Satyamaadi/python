```
import cv2
import numpy as np
```

```
!wget https://raw.githubusercontent.com/Apress/building-computer-vision-apps-artificial-neural-networks/master/chapter3/images/zebra.png
```

```
image = cv2.imread('zebra.png')
```

```
type(image)
```

```
(h,w) = image.shape[:2]
```

```
aspect = w /h
```

```
height = int(0.5 * h)
```

```
width = int(height * aspect)
```

```
dimension = (width,height)
```

```
resizedImage = cv2.resize(image,dimension,interpolation=cv2.INTER_AREA)
```

```
import matplotlib.pyplot as plt
```

```
plt.imshow(image)
```

```
plt.imshow(resizedImage)
```
