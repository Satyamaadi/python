```
import cv2
import numpy as np
```

```
!wget https://raw.githubusercontent.com/Apress/building-computer-vision-apps-artificial-neural-networks/master/chapter3/images/zebrasmall.png
```

```
image = cv2.imread('zebrasmall.png')
```

```
type(image)
```

```
(h,w) = image.shape[:2]
```

```
center = (h//2,w//2)
```

```
angle = -45
```

```
scale  = 1.0
```

```
rotation_matrix = cv2.getRotationMatrix2D(center,angle,scale)
```

```
rotated_image = cv2.warpAffine(image,rotation_matrix,(image.shape[1],image.shape[0]))
```

```
import matplotlib.pyplot as plt
```

```
plt.imshow(image)
```

```
plt.imshow(rotated_image)
```
